In this exercise, you will scrap data from Wikipedia. More specifically, you will design a code that lets you capture the population of the seven Scottish cities

You will scrap the following Wiki entry, making sure that you only capture data from the cities!

https://en.wikipedia.org/wiki/List_of_towns_and_cities_in_Scotland_by_population

First, I recommend you to copy-paste the code from the lecture

When you want to find the table, keep in mind the class of the table! You can do so by specifying it as:

    rows = soup.find("table", {"class":"class_of_the_table"})

```python
# create two lists to put names and populations
cities = []
population = []
# define the url
url = "https://en.wikipedia.org/wiki/List_of_towns_and_cities_in_Scotland_by_population"
# request the url
r = requests.get(url)
# soup it
soup = BeautifulSoup(r.content,"html.parser")
# get the rows of the table
rows = soup.find("table", {"class":"wikitable sortable"}).find_all("tr")
for row in rows:
    cells = row.find_all("td")
    print(cells)
    if cells and "City" in cells[2].get_text(): # if city, then get name and pop
        cities.append(cells[0].get_text().split("\n")[0])
        population.append(int(cells[1].get_text().split("\\")[0].replace(",","")))
print(cities,population)
```

In the end, you should have a data frame that looks something like this:

```python
import pandas as pd
scotland = pd.DataFrame({'City':cities,'Population':population})
scotland
```
