```python
# Importing the necessary files
import pandas as pd
# This code makes charts appear in the notebook in the first run
%matplotlib inline
```

```python
df = pd.read_csv('https://www.dropbox.com/scl/fi/b4vyh559u88j8jlzkhwz1/city_populations_updated.csv?rlkey=ak3juioaq506kyt3d3qh7cr1s&raw=1')
df
```

```python
# Only include years above a certain limit (otherwise takes ages!)
df2 = df[df['Year']>1991]
df2.reset_index()
```

```python
# Pivot data
df3 = df2.pivot(index='Year',
                          columns='City',
                          values=['Population'])
df3 = df3['Population']# If we don't do this, then the word population will appear next to the city in the BCR
df3
```

```python
import bar_chart_race as bcr
bcr.bar_chart_race(df=df3, filename=None,
                   n_bars=12, # if I use 12, then weird things happen
                   perpendicular_bar_func='median',
                   title='Most Populous Cities in the World')
```
